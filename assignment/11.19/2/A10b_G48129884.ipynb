```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

Import data.

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv",)
```

Drop colunmns which have null values to reduce dimension.

```python
train.dropna(axis = 1,inplace=True)
test.dropna(axis = 1,inplace=True)
```

Let's take a look at the data.

```python
train.head()
```

```python
test.head()
```

the number of columns.

```python
len(list(train.columns))
```

```python
test_id = test.ID
```

Y is the target varible

```python
Y = train.target
```

```python
train = train.ix[:,1:-1]
```

```python
test = test.ix[:,1:]
```

Drop all categorical columns.

```python
temp = (train.dtypes == "int64")
```

```python
col = pd.DataFrame(data = list(train.columns))[list(temp)][0].tolist()
```

```python
temp = (train.dtypes == "float64")
```

```python
col = col + pd.DataFrame(data = list(train.columns))[list(temp)][0].tolist()
```

```python
col
```

```python
train = train[col]
test = test[col]
```

```python
test.head()
```

Fit data into classfication model.

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(train,Y)
```

Predict value.

```python
y_pred = classifier.predict(test)
```

```python
y_pred = pd.Series(y_pred)
```

```python
test_id =test_id.to_frame(name ="ID")
```

```python
y_pred = y_pred.to_frame(name = "target")
```

```python
pred = pd.merge(test_id,y_pred, left_index=True, right_index=True) 
```

Output to csv file.

```python
pred.to_csv("pred.csv",index= False)
```
